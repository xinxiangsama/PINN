```python
import scipy.integrate as syint
import numpy as np
import matplotlib.pyplot as plt

# 参数定义
sigma = 10
beta = 8/3
rhos = [29, 28.8]
results = []

for rho in rhos:
    # 定义微分方程形式
    def diff3(d_list, t):
        x, y, z = d_list
        return np.array([sigma*(y-x), rho*x-y-x*z, x*y-beta*z])

    t = np.linspace(0, 100, 100000)
    # 微分方程求解
    result = syint.odeint(diff3, [1, 1, 1], t)
    results.append(result)

# 绘制图表
fig = plt.figure()
ax = fig.add_subplot( 111,projection='3d')


for rho, result in zip(rhos, results):
    ax.plot(result[:, 0], result[:, 1], result[:, 2], label=f'rho={rho}')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Lorenz equations with different rho values ')
ax.legend()

plt.show()
```
